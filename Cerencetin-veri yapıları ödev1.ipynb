5.SORU

```python
ifade= "Hi-Kod Veri Bilimi Atölyesi "
kelimeler=ifade.split()
buyuk_harf=ifade.upper()
kucuk_harf=ifade.lower()
sayilar="0123456789"
ciftsayilar="".join([sayi for sayi in sayilar if int(sayi)%2==0])
teksayilar="".join([sayi for sayi in sayilar if int(sayi)%2!=0 ])
print("Kelime listesi:",kelimeler)
print("Büyük harfli ifade:",buyuk_harf)
print("Küçük harfli ifade:",kucuk_harf)
print("Çift sayılar:",ciftsayilar)
print("Tek sayılar:",teksayilar)
```
